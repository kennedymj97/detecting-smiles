```python
import caffe2.python.onnx.backend as backend
import numpy as np
import onnx
import cv2
import PIL
from torchvision import transforms
from fastai.vision import normalize, imagenet_stats
import torch
```

```python
def image_loader(image_name, imsize=64):
    """load image, returns (cuda) tensor"""
    loader = transforms.Compose([transforms.Resize(imsize), transforms.ToTensor()])
    image = cv2.cvtColor(image_name, cv2.COLOR_BGR2RGB)
    image = PIL.Image.fromarray(image) 
    image = loader(image).float()
    image.clone().detach().requires_grad_(True)
    image = image.unsqueeze(0)
    return image.numpy()  # add .cuda() if using gpu
```

```python
def preprocess_img(img):
    """Prepare image to be used by caffe2 model"""
    img = img / 255
    img = np.rollaxis(img, 2, 0)
    mn,std = imagenet_stats
    img = normalize(torch.FloatTensor(img),torch.FloatTensor(mn),torch.FloatTensor(std))
    img = img.unsqueeze(0)
    return img.numpy()
```

```python
def predict(model, input_image):
    outputs = model.run(input_image)
    result = np.argmax(outputs[0])
    if result:
        return "Not happy"
    else:
        return "Happy :)"
```

```python
def softmax(x):
    """Compute softmax values for each sets of scores in x."""
    e_x = np.exp(x - np.max(x,axis=1,keepdims=True))
    return e_x / e_x.sum(axis=1,keepdims=True)
```

```python
def preprocess_img(img):
    """Prepare image to be used by caffe2 model"""
    img = cv2.resize(img, (64, 64))
    img = img / 255
    img = np.rollaxis(img, 2, 0)
    mn,std = imagenet_stats
    img = normalize(torch.FloatTensor(img),torch.FloatTensor(mn),torch.FloatTensor(std))
#     img = img.unsqueeze(0)
    img = img.numpy()
    return np.expand_dims(img, axis=0)
```

```python
def predict(model, input_image):
    outputs = model.run(input_image)
    result = np.argmax(outputs[0])
    probs = softmax(outputs[0]) * 100
    if result:
        return "Not happy  {:.1f}%".format(probs[0][1])
    else:
        return "Happy :)  {:.1f}%".format(probs[0][0])
```

```python
onnx_model = onnx.load('./onnx_models/happy.onnx')
caffe_resnet18 = backend.prepare(onnx_model, device='CPU')
```
